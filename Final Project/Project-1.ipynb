```python
import json
import requests
from config import api_key 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from mpl_toolkits.axes_grid1 import host_subplot
import numpy as np
```

```python

gasPrices = "Data/GasPrices.csv"


gasPrices = pd.read_csv(gasPrices)
gasPrices= gasPrices.rename(columns={"Boston, MA Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)": "Boston", "Chicago Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)": "Chicago", "Cleveland, OH Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)": "Cleveland","Denver Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)": "Denver", 
                                        "Houston Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)":"Houston",
                                     "Los Angeles Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)":"Los Angeles", 
                                     "Miami, FL Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)":"Miami", 
                                     "New York City Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)":"New York", 
                                     "San Francisco Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)":"San Francisco", 
                                     "Seattle, WA Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)":"Seattle"})

gasPrices.melt(id_vars=["Date"], 
        var_name="City", 
        value_name="Value")

gasPrices2019 = gasPrices[gasPrices['Date'].str.contains("2019")]


gasPrices2019 = gasPrices2019.replace(['Jan-2019','Feb-2019',
'Mar-2019',
'Apr-2019',
'May-2019',
'Jun-2019',
'Jul-2019',
'Aug-2019',
'Sep-2019',
'Oct-2019',
'Nov-2019',
'Dec-2019'],['01-January',
'02-February',
'03-March',
'04-April',
'05-May',
'06-June',
'07-July',
'08-August',
'09-September',
'10-October',
'11-November',
'12-December',])



Gas = gasPrices2019.melt(id_vars=["Date"], 
        var_name="City", 
        value_name="Gas Price")

Gas["Unique"] = Gas["Date"] + Gas["City"]

Gas
```

```python
Temps = "Data/Avg Temp per Month.csv"

Temps = pd.read_csv(Temps)

Temps = Temps.drop(['Unnamed: 0'], axis=1)

Temps = Temps.melt(id_vars=["City"], 
        var_name="Date", 
        value_name="Weather")

Temps = Temps.replace(['January','February',
'March',
'April',
'May',
'June',
'July',
'August',
'September',
'October',
'November',
'December'],['01-January',
'02-February',
'03-March',
'04-April',
'05-May',
'06-June',
'07-July',
'08-August',
'09-September',
'10-October',
'11-November',
'12-December',])

Temps = Temps[['Date', 'City', 'Weather']]
Temps["Unique"] = Temps["Date"] + Temps["City"]
Temps
```

```python
#Merge Tables
Combined = pd.merge(left=Temps, right=Gas, how='left', left_on='Unique', right_on='Unique')

# Rename Columns
Combined = Combined.rename(columns={"Date_x": "Date", 
                         "City_x": "City"})
# Drop Columns
Combined = Combined.drop(['Unique','Date_y','City_y'], axis=1)
Combined
```

```python
Date_Array=Combined['Date'].unique()
Avg_Month=Combined.groupby(['Date']).mean()
Gas_Price_Date = Avg_Month['Gas Price'].tolist()
Weather_avg_Date = Avg_Month['Weather'].tolist()
City_Array=Combined['City'].unique()
Avg_City=Combined.groupby(['City']).mean()
Gas_Price_City = Avg_City['Gas Price'].tolist()
Weather_City=Avg_City['Weather'].tolist()
```

```python

plt.style.use('ggplot')

x = Date_Array
G_Price = Gas_Price_Date

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, G_Price, color='green')
plt.xlabel("Month - 2019")
plt.ylabel("Average Gas Price")
plt.title("Average Gas Prices Per Month in 2019")

plt.xticks(x_pos, x,rotation=90 )


plt.savefig("Graphs/Gas_Month.PNG")
```

```python


plt.style.use('ggplot')

x = City_Array
G_Price_City = Gas_Price_City

x_pos = [i for i, _ in enumerate(x)]
x_pos

plt.bar(x_pos, G_Price_City, color='blue')
plt.xlabel("City")
plt.ylabel("Average Gas Price")
plt.title("Average Gas Prices Per City in 2019")

plt.xticks(x_pos, x,rotation=90 )

plt.savefig("Graphs/Gas_City.PNG")
```

```python
plt.scatter(Combined["Weather"], Combined["Gas Price"], marker = "o")
slope, intercept, r_value, p_value, stderr = stats.linregress(Combined["Weather"], Combined["Gas Price"])
linear = slope * Combined["Weather"] + intercept

line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))
plt.annotate(line_eq, (21, 3.75), fontsize = 14, color = "red")
plt.plot(Combined["Weather"], linear, "-r")
plt.xlabel("Avg Temp (Â°F)")
plt.ylabel("Avg Gas Price ($)")
plt.title("Avg Gas Prices ($) vs. Avg Temprature (Â°F)")
plt.savefig("Graphs/Scatter Plot.png")
```

```python
correlation = stats.pearsonr(Combined["Weather"], Combined["Gas Price"])
if correlation[0] > 0 < .5:
    print(f"The correlation coefficient between temperature and gas prices is {round(correlation[0],2)} and shows that there is a weak correlaiton")
elif correlation > .5 <.8:
    print(f"The correlation coefficient between temperature and gas prices is {round(correlation[0],2)} and shows that there is a moderate correlation")
elif correlation > .8:
    print(f"The correlation coefficient between temperature and gas prices is {round(correlation[0],2)} and shows that there is a strong correlation")
```
