```python
import json
import requests
from config import api_key #for crime data
import pandas as pd
```

```python
# Study data files
Census_IDs = "Census_IDS.csv"
Locations = "Top 50 cities.csv"

# Read the mouse data and the study results
Census_IDs = pd.read_csv(Census_IDs)
Locations = pd.read_csv(Locations)
```

```python
#API details for Census
url = "https://api.census.gov/data/2019/pep/population?get=POP,NAME&for=place:00124&in=state:01"


# Build query URL
query_url = url
```

```python
# Get Census data
census_response = requests.get(query_url)
census_json = census_response.json()

# Print Census Data
print(census_json)
```

```python
cities = ["Paris", "London", "Oslo", "Beijing"]

# set up lists to hold reponse info
lat = []
temp = []

# Loop through the list of cities and perform a request for data on each
for city in cities:
    response = requests.get(query_url + city).json()
    lat.append(response['coord']['lat'])
    temp.append(response['main']['temp'])

print(f"The latitude information received is: {lat}")
print(f"The temperature information received is: {temp}")
```
